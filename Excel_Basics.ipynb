1.	Explain the difference between Relative, Absolute and Mixed Cell Referencing.
You have the following dataset:
Vehicle	Price
Car	500000
Bicycle	10500
Auto-Rickshaw	25000

Ans: Relative Referencing: Ensures formulas adapt automatically

Absolute Referencing: Prevents errors when you always need the same value. Useful for fixed values. Written with the $ sign


Mixed cell referencing: Gives flexibility in Row/Column operations. It is the combination of relative and absolute cell referencing.
Without correct referencing copied formula may gives wrong answers.
